import library

```python
import cv2
import os
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import easyocr
```



---


open data. Because my data in drive, so i should write the syntax like this



```python
reader = easyocr.Reader(['en'])
train_path = '/content/drive/My Drive/capstone/train/'
val_path = '/content/drive/My Drive/capstone/val/'
```



---


read dataset, extract one image from dataset with easyocr, and cleaning the data to be table format


```python
def extract_and_clean(image_path):
    result = reader.readtext(image_path, detail=0)

    raw_text = " ".join(result).lower()

    cleaned_data = {}
    for line in raw_text.split("\n"):
        if any(char.isdigit() for char in line):
            key_value = line.split(":") if ":" in line else line.split()
            if len(key_value) >= 2:
                key = key_value[0].strip()
                value = key_value[1].strip().replace(",", ".").replace("%", "").replace(" ", "")
                cleaned_data[key] = value

    return cleaned_data
```



---


apply extract_and_clean one data to one dataset and save in directory

```python
def process_folder(folder_path):
    data = []
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if file_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            extracted_data = extract_and_clean(file_path)
            extracted_data['file_name'] = file_name
            data.append(extracted_data)
    return pd.DataFrame(data)

train_data = process_folder(train_path)
val_data = process_folder(val_path)
```



---


in here im save extracted data in .csv format so the data easy to use

```python
train_data.to_csv('/content/train_data_cleaned.csv', index=False)
val_data.to_csv('/content/val_data_cleaned.csv', index=False)
```
